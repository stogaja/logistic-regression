## Importing the dataset

```python
# lets import pandas and numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# let's read the dataset
dataset = pd.read_csv('Data.csv')
```

```python
# lets preview the dataset
dataset.head()
```

```python
# lets split the dataset into x and y using iloc
X = dataset.iloc[:, :-1].values # selects all rows but the target column which is the last column
y = dataset.iloc[:,-1]  # select only the last column as it contains our labels
```

## Splitting the dataset into the Training set and Test set

```python
# let's import train test split from sklearn
from sklearn.model_selection import train_test_split

# lets split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training CatBoost on the Training set

```python
%pip install catboost
```

```python
# let's import the catboost classifier
from catboost import CatBoostClassifier

classifier = CatBoostClassifier()
classifier.fit(X_train, y_train)
```

## Making the Confusion Matrix

```python
# let's make a confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score

# defining y_pred
y_pred = classifier.predict(X_test)

# defining the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# printing the confusion matrix
print(cm)

accuracy_score(y_test, y_pred)

# let's print the accuracy score as a percentage
print(f'The accuracy score = {round(accuracy_score(y_test, y_pred)* 100)}%')

# alternatively you can assign the accuracy score to a variable then use the variable in the formatted string
```

## Applying k-Fold Cross Validation

```python
# let's apply k-folds cross validation
# 1. we import the cross validation score
from sklearn.model_selection import cross_val_score
# 2. We create a new object of KFold class and pass it with number of fold as parameter, in this case is set to three
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)

# 3. printing the accuracy output
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))

# 4. printing the standard deviation
print('Standard Deviation: {:.2f} %'.format(accuracies.std()*100))
```

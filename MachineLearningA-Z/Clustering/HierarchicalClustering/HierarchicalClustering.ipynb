## Hierarchial Clustering Omdena

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, [3, 4]].values
```

```python
dataset.head(10)
```

```python
# produce dendogram and find optimal number of clusters
# ward method used to minimize variance within clusters
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method = "ward"))
plt.title("Dendrogram")
plt.xlabel("Observation Points")
plt.ylabel("Euclidean Distance")
```

```python
# train the hierarchical clustering model on the dataset
# using 3 or 5 clusters is a good idea based on the dendro
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity = "euclidean", linkage = "ward")

# use the fit predict method
y_hc = hc.fit_predict(x)
```

```python
# print clusters
print(y_hc)
```

```python
# visualise the clusters
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 50, c = "red", label="Clus. 1")
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 50, c = "blue", label="Clus. 2") 
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], s = 50, c = "green", label="Clus. 3") 
plt.scatter(x[y_hc == 3, 0], x[y_hc == 3, 1], s = 50, c = "cyan", label="Clus. 4")
plt.scatter(x[y_hc == 4, 0], x[y_hc == 4, 1], s = 50, c = "magenta", label="Clus. 5")

plt.title("Clusters of Customers")
plt.xlabel("Income")
plt.ylabel("Spending Score 1-100")
plt.legend()
plt.show()
```
